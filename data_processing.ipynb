```python
df = pd.DataFrame(columns=[
    'uid',
    'date',
    'level',
    # 'danmu_count',
    'yc_gift_value',
    'visit_count',
])
df
# df.to_excel('data/detail_400_audiences_2022-01-01_2022-06-15.xlsx')
```

```python
data_path = 'data/detail_350_audiences_2022-01-01_2022-06-15.jsonl'
test_path = 'data/testdata.jsonl'

with open(data_path, 'r', encoding='utf-8') as fin:
    for i, line in enumerate(fin):
        item = json.loads(line)
        if item['hasData']:
            df = df.append(
                [{
                    'uid': item['uid'],
                    'date': item['date'],
                    'level': item['level'],
                    'visit_count': float(item['totalItemVo']['visitCount']),
                    'yc_gift_value': float(item['totalItemVo']['yc_gift_value'])
                }])
        else:
            df = df.append(
                [{
                    'uid': item['uid'],
                    'date': item['date'],
                    'level': item['level'],
                    'visit_count': 0.0,
                    'yc_gift_value': 0.0
                }])
```

```python
df.tail()
```

```python
uids = list(set(df['uid']))
uids
```

```python
len(df)
```

```python
df2 = pd.DataFrame(columns=['uid', 'level'])
df2['uid'] = uids
df2.set_index('uid', inplace=True)
df2
```

```python
# df[df['uid'] == 357558595]
"""
第 i 周 打赏次数 打赏金额
"""
for uid in uids:
    tmp = df[df['uid'] == uid]
    df2.loc[uid]['level'] = int(tmp.iloc[0]['level'])
    # assert len(tmp) == 166, '{}'.format(uid)
    for i in range(len(tmp) // 7):
        week_df = tmp.iloc[7 * i:7 *(i + 1)]
        week_index = ' week {} ('.format(i) + week_df.iloc[0]['date'] + '_' + week_df.iloc[-1]['date'] + ')'
        df2.loc[uid, 'gift_value' + week_index] = week_df['yc_gift_value'].sum()
        df2.loc[uid, 'gift_times' + week_index] = (week_df['yc_gift_value'] > 0).sum()
```

```python
df2
```

```python
df3 = pd.DataFrame(columns=[
    'uid',
    'month',
    'level',
    # 'danmu_count',
    'yc_gift_value',
    'visit_count',
])

with open('data/detail_350_audiences_2022-01_2022-06.jsonl', 'r', encoding='utf-8') as fin:
    for line in fin:
        item = json.loads(line)
        if item == {}:
            continue
        if item['hasData']:
            df3 = df3.append(
                [{
                    'uid': item['uid'],
                    'month': item['month'],
                    'level': item['level'],
                    'visit_count': float(item['totalItemVo']['visitCount']),
                    'yc_gift_value': float(item['totalItemVo']['yc_gift_value'])
                }])
        else:
            df3 = df3.append(
                [{
                    'uid': item['uid'],
                    'month': item['month'],
                    'level': item['level'],
                    'visit_count': 0.0,
                    'yc_gift_value': 0.0
                }])
```

```python
df3
```

```python
for uid in uids:
    tmp = df3[df3['uid'] == uid]
    df2.loc[uid]['level'] = int(tmp.iloc[0]['level'])
    for i in range(len(tmp)):
        df2.loc[uid, 'gift_value ' + tmp.iloc[i]['month']] = tmp.iloc[i]['yc_gift_value']
        df2.loc[uid, 'visit_count ' + tmp.iloc[i]['month']] = tmp.iloc[i]['visit_count']
```
